# 📘 Python – Conditional Statements

## 🔵 1️⃣ What Are Conditional Statements?
- Conditional statements let Python **decide which block of code to execute** based on conditions.
- Uses **Boolean logic** (`True` or `False`).

✅ **Example:**
```python
age = 18
if age >= 18:
    print("You are an adult")

🔵 2️⃣ Types of Conditional Statements in Python
1️⃣ if statement
2️⃣ if-else statement
3️⃣ if-elif-else ladder
4️⃣ Nested if statements
5️⃣ Ternary conditional operator

🔵 3️⃣ Basic if Statement

if condition:
    # code block

temperature = 30
if temperature > 25:
    print("It's hot outside!")

🔵 4️⃣ if-else Statement

if condition:
    # code if condition is true
else:
    # code if condition is false


age = 17
if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")

🔵 5️⃣ if-elif-else Ladder
✅ Used when there are multiple conditions to check.

if condition1:
    # code block 1
elif condition2:
    # code block 2
else:
    # code block 3


marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")


🔵 6️⃣ Nested if Statements
✅ You can write an if statement inside another if statement.


num = 10
if num > 0:
    if num % 2 == 0:
        print("Positive Even Number")
    else:
        print("Positive Odd Number")
else:
    print("Negative Number")


🔵 7️⃣ Ternary Conditional Operator (Inline if-else)
✅ Short-hand way to write if-else in one line.

value_if_true if condition else value_if_false


age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)  # Adult


🔵 8️⃣ Conditional Statements with Logical Operators
✅ Combine multiple conditions using and, or, not.

age = 25
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")
else:
    print("Entry denied")


🔵 9️⃣ Conditional Statements with Membership & Identity
✅ You can also use in, not in, is, is not.

fruits = ["apple", "banana", "cherry"]

if "banana" in fruits:
    print("Banana is available")

x = None
if x is None:
    print("x has no value")

🔵 🔟 Special Cases & Best Practices
✅ Pass Statement – Used when you want an empty block.

if True:
    pass  # Do nothing


✅ Avoid Deep Nesting

# ❌ Hard to read
if a:
    if b:
        if c:
            print("Too nested")

# ✅ Better with elif or functions
if a and b and c:
    print("Clean and readable")



✅ Use meaningful conditions

# ❌ Bad
if x == True:
    ...

# ✅ Better
if x:
    ...
