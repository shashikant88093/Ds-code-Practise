## ğŸ”µ 2ï¸âƒ£ for Loop
âœ… Used to iterate over **sequences** like list, tuple, set, dictionary, or string.

âœ… **Basic Syntax:**
```python
for item in sequence:
    # code block


fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)


âœ… Using range() with for loop
range(start, stop, step) â€“ generates a sequence of numbers.

for i in range(1, 6):
    print(i)


âœ… Iterating with Index (using enumerate())


colors = ["red", "green", "blue"]
for index, color in enumerate(colors):
    print(index, color)


ğŸ”µ 3ï¸âƒ£ while Loop
âœ… Executes code as long as the condition is True.

while condition:
    # code block


count = 1
while count <= 5:
    print(count)
    count += 1


ğŸ”µ 4ï¸âƒ£ Loop with else
âœ… Python allows an else block with for and while loops.
âœ… The else block runs only if the loop finishes normally (not stopped by break).

for i in range(3):
    print(i)
else:
    print("Loop Finished!")

for i in range(3):
    print(i)
else:
    print("Loop Finished!")


ğŸ”µ 5ï¸âƒ£ Jump Statements in Loops
âœ… break â†’ Exit the loop immediately
âœ… continue â†’ Skip the current iteration & go to the next
âœ… pass â†’ Does nothing (used as placeholder)

âœ… break Example

for i in range(1, 6):
    if i == 3:
        break
    print(i)


for i in range(1, 6):
    if i == 3:
        continue
    print(i)


for i in range(3):
    pass  # Placeholder â€“ no operation performed



ğŸ”µ 6ï¸âƒ£ Nested Loops

for i in range(1, 4):
    for j in range(1, 4):
        print(f"({i},{j})", end=" ")
    print()


ğŸ”µ 7ï¸âƒ£ Useful Loop Functions
âœ… range() â€“ for number sequences
âœ… enumerate() â€“ get index + value
âœ… zip() â€“ iterate over multiple sequences at once

names = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 92]

for name, score in zip(names, scores):
    print(name, score)


ğŸ”µ 8ï¸âƒ£ Infinite Loops
âœ… A while True loop runs forever unless break is used.


while True:
    print("This runs forever until break")
    break


ğŸ”µ 9ï¸âƒ£ Advanced Looping
âœ… List Comprehension (Compact looping)

squares = [x**2 for x in range(1, 6)]
print(squares)  # [1, 4, 9, 16, 25]


âœ… Dictionary Comprehension

numbers = {x: x**2 for x in range(1, 4)}
print(numbers)  # {1: 1, 2: 4, 3: 9}


âœ… Using else with while loop

i = 0
while i < 3:
    print(i)
    i += 1
else:
    print("Done with while loop")
